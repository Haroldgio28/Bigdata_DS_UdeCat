## **Calificaciones**

Escribe un programa que solicite una puntuación entre 0.0 y 1.0. Si la
puntuación está fuera de ese rango, muestra un mensaje de error. Si la puntuación
está entre 0.0 y 1.0, muestra la calificación usando la tabla siguiente:


| Puntuación | Calificación |
| ---------| -----------|
|>= 0.9 |Sobresaliente|
|>= 0.8| Notable|
|>= 0.7 |Bien|
|>= 0.6 |Suficiente|
|< 0.6 |Insuficiente|

```
Introduzca puntuación: 0.95
Sobresaliente
Introduzca puntuación: perfecto
Puntuación incorrecta
Introduzca puntuación: 10.0
Puntuación incorrecta
Introduzca puntuación: 0.75
Bien
Introduzca puntuación: 0.5
Insuficiente
```


## **Salario**

Escriba una función que solicite número de horas trabajadas a la semana y retorne el salario del trabajador. Si el trabajador laboró por 40 horas o menos, el pago por hora será de 15.000. Para las horas extra trabajadas (más de 40 horas a la semana) el pago por hora será de 21.000.

```
Input: 35
Out: 'Su pago es de: 525000 pesos'

Input: 40
Out: 'Su pago es de: 600000 pesos'

Input: 48
Out: 'Su pago es de: 768000 pesos'
```



## **La secuencia Collatz**

Escriba una función llamada ``collatz()`` que tenga un solo parámetro llamado ``número``.

Si ``número`` es par, ``collatz()`` debe imprimir ``número // 2`` y retornar este valor.

Si ``número`` es impar, ``collatz()`` debe imprimir y retornar ``3 * número + 1`` .

Ahora escriba un programa que le pida al usuario un número entero, y que llame a la función  ``collatz()`` tantas veces como sea necesario hasta que la salida sea 1. Tarde o temprano, para cualquier número entero la salida siempre será 1.

La salida de este programa se ve más o menos así.

```
In: 6
Out:
3
10
5
16
8
4
2
1
```
